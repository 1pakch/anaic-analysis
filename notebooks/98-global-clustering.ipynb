```r
library(purrr)
library(tidyverse)
library(Seurat)
library(ggplot2)
library(scales)
```

```r
source('paths.R')
source('factors.R')
source('misc.R')
source('plotting.R')
```

# 1. Load Data

```r
d <- load_object(path_at('out', 'cache')('annotated', 'Robj'))
```

# 2. SCTransform

```r
# SCT data is not properly reconstructed when persisted hence recomputation

d <- (d %>% SCTransform(verbose=FALSE, verbosity=0))
```

# 3. Clustering

## 3.A Load Hallmark Genes

```r
hallmark.genes.colspec <- cols(
  gene = col_character(),
  cell_type = col_character(),
  cell_subtype = col_character(),
  essential = col_logical()
)

hallmark.genes <- (
    read_csv('./in/hallmark_genes/hallmark_all.csv', col_types = hallmark.genes.colspec)
    %>% mutate(gene=toupper(gene))
    %>% filter(gene != 'FUT4')
    %>% filter(gene != 'TFF3')
)
```

```r
stopifnot(all(hallmark.genes$gene %in% rownames(d)))
```

## 3.B Clustering itself

```r
cluster.dims <- 10

d <- (
    d
    %>% FindNeighbors(dims=NULL, features=hallmark.genes$gene, k.param=31, verbose=FALSE)
    %>% FindClusters(resolution=0.25, verbose=FALSE)
)
```

```r
# Check that the class counts are the exactly the same when the annotation was performed and
# raises an error otherwise (pointless to run the code below if the clustering has changed).
#
# Note that the clustering is somewhat unstable, e.g. depends on exact package versions etc.

seurat_cluster_counts <- (d@meta.data %>% group_by(seurat_clusters) %>% count() %>% pull(n))
all(seurat_cluster_counts == c(4513, 2908, 2820, 1694, 843)) %>% stopifnot()
```

## 3.C Labeling of clusters

```r
cluster_labels_df <- (
    left_join(
        d %>% get_metadata() %>% select(barcode, seurat_clusters),
        global_clusters_meta,
        by = 'seurat_clusters')
    %>% select(-seurat_clusters)
)
```

## 3.D Plot Clusters

```r
options(repr.plot.width=7, repr.plot.height=6)
d <- d %>% RunPCA(verbose=FALSE) %>% RunTSNE(dims=1:15, verbose=FALSE, reduction.name='tsne_pcs')
Seurat::DimPlot(d, group.by='global_clusters', reduction = 'tsne_pcs') + scale_color_manual(values=global_cluster_colors)
```

# 4. Save Clusters

```r
write_csv(cluster_labels_df, path_at('out', 'precomputed')('global_clusters_and_cell_types.csv'))
```
