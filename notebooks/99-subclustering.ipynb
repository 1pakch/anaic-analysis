# 2. Find differential genes btw conditions

```r
diff_genes_1_vs_rest <- function(data, by, of, ...) {
    FindMarkers(data, group.by=by, ident.1=of, ...)
}
```

```r
# Differential genes 1 group versus the others
find.diff.genes <- function(data, by, only.pos=TRUE, p_val_adj_max=0.01, min.pct=0.1, logfc.threshold=0.25) (
    unique(data@meta.data[[by]])
    %>% {setNames(., .)}
    %>% map(~diff_genes_1_vs_rest(data, by, .x, only.pos=only.pos, min.pct=min.pct, logfc.threshold=logfc.threshold))
    %>% map(~tibble::as_tibble(.x, rownames='gene'))
    %>% imap(~mutate(.x, group=.y))
    %>% bind_rows
    %>% filter(p_val_adj <= p_val_adj_max) 
    %>% group_by(gene)
    %>% slice_min(p_val_adj, n=1)
    %>% slice_head() # if minimum not unique
    %>% ungroup()
    %>% mutate(group=factor(group, levels=levels(data@meta.data[[by]])))
    %>% arrange(group, p_val_adj)
    %>% rename(!!by := group)
)
```

```r
diff.genes.conditions <- (
    split
    %>% map(find.diff.genes, 'condition_text')
)
```

# 3. Subclustering

## 3.1 PCA on diff genes

```r
# Use exact SVD instead of approximation used by Seurat by default

split <- (
    split
    %>% imap(~RunPCA(.x, assay='SCT', verbose=FALSE, features=diff.genes.conditions[[.y]]$gene, approx=FALSE))
)
```

```r
options(repr.plot.width=9, repr.plot.height=6)
(split %>% map(ElbowPlot, ndim=32) %>% imap(~.x + ggtitle(.y) + ylim(0, 30)) %>% reduce(`+`))
```

## 3.2 Clustering itself

```r
cluster.dims <- 10
split <- (
    split
    %>% imap(~FindNeighbors(.x, dims=1:cluster.dims, k.param=31, verbose=FALSE))
    %>% imap(~FindClusters(.x, resolution=0.35, verbose=FALSE))
    %>% imap(~RunUMAP(.x, dims=1:cluster.dims, n.neighbors=31, seed=10, verbose=FALSE))
)
```

## 3.3 Reproducibility Check

```r
# Check that the class counts are the exactly the same when the annotation was performed and
# raise an error otherwise (pointless to run the code below if the clustering has changed).
#
# Note that the clustering is somewhat unstable, e.g. depends on exact package versions etc.

expected_cluster_counts <- list(
    Basal = c(1180, 1019, 824, 565, 544, 381),
    BdiS = c(557, 358, 257, 213, 181, 128),
    Secretory = c(1062, 848, 525, 385),
    Ciliated = c(1406, 823, 392, 150, 137)
)

compute_cluster_counts <- function(obj) (
    obj
    %>% get_metadata
    %>% group_by(seurat_clusters)
    %>% count()
    %>% pull(n)
)

(
    split
    %>% map(compute_cluster_counts)
    %>% map2(expected_cluster_counts, ~all(.x == .y))
    %>% map(stopifnot)
) -> .
```

## 3.4 Cluster labelling

```r
# Ciliated cluster 4 is like Undefined cell type, drop it

split$Ciliated <- split$Ciliated[,split$Ciliated@meta.data$seurat_clusters != '4']
```

```r
labels_from_one <- function(f) {
    labels <- (1:(f %>% levels %>% length)) %>% as.character()
    indices <- (f %>% as.integer)
    factor(labels[indices], levels=labels, labels=labels)
}

subcluster_orders <- list(
    Basal = c('1', '3', '4', '6', '2', '5'),
    BdiS = c('1', '2', '3', '4', '5', '6'),
    Secretory = c('1', '2', '5', '4', '3'),
    Ciliated = c('1', '2', '3', '4')
)

add_prefix <- function(x, y) {
    fct_relabel(x, ~paste(substr(y, 1, 3), .x, sep=''))
}

subclusters <- (
    split
    %>% map(~.x@meta.data %>% pull(seurat_clusters))
    %>% map(labels_from_one)
    %>% imap(~factor(.x, levels=subcluster_orders[[.y]]))
)

subclusters$Secretory <- (
    subclusters$Secretory
    %>% fct_recode(`1`='1', `1`='2', `1`='5', `2`='4', `3`='3')
    %>% fct_drop()
)

split <- (
    split
    %>% imap(~AddMetaData(.x, subclusters[[.y]], 'subcluster'))
)
```

```r
add_subcluster2 <- function(obj, prefix) {
    subcluster2 <- factor(paste(prefix, as.integer(obj@meta.data$subcluster), sep=''))
    AddMetaData(obj, subcluster2, 'subcluster2')
}

reorder_subcluster2 <- function(obj, order) {
    subcluster2 <- obj@meta.data$subcluster2
    subcluster2 <- factor(as.character(subcluster2), levels=order)
    AddMetaData(obj, subcluster2, 'subcluster2')
}

add_subcluster2_text <- function(obj, extras) {
    values <- left_join(
        obj@meta.data %>% select(subcluster2),
        extras %>% select(subcluster2, labels))$labels
    levels <- extras %>% arrange(order) %>% pull(labels)
    fct <- factor(values, levels=levels, ordered=TRUE)
    AddMetaData(obj, fct, 'subcluster2_text')
}

prefixes <- list(Basal='b', `BdiS`='bs', Secretory='s', Ciliated='c')

orders <- list(
    Basal = c('b1', 'b2', 'b4', 'b3', 'b5', 'b6'),
    BdiS = c('bs1', 'bs2', 'bs3', 'bs4', 'bs5', 'bs6'),
    Ciliated = c('c1', 'c2', 'c3', 'c4'),
    Secretory = c('s2', 's1', 's3')
)

extras <- list(
    Basal = data.frame(
        subcluster2 = c('b1', 'b2', 'b3', 'b4', 'b5', 'b6'),
        order = c(1, 2, 5, 3, 4, 6),
        color = c('#00ccff', '#6699cc', '#99ff99', '#9999ff', '#66cc00', '#ffcccc'),
        labels = c(
            'Steady state', 'Differentiating', 'Highly proliferating',
            'Diff. KRT14/KRT16 high', 'Low proliferating', 'Inflamed'
        )
    ),
    Ciliated = data.frame(
        subcluster2 = c('c1', 'c2', 'c3', 'c4'),
        order = c(1, 2, 3, 4),
        color = c('#00ccff', '#6699cc', '#ffcccc', '#ff0033'),
        labels = c(
            'Steady state', 'Immature', 'Inflamed IFN-\U03BB\U2212', 'Inflamed IFN-\U03BB+'
        )
    ),   
    BdiS = data.frame(
        subcluster2 = c('bs1', 'bs2', 'bs3', 'bs4', 'bs5', 'bs6'),
        order = c(1, 2, 4, 3, 6, 5),
        color = c('#00ccff', '#6699cc', '#99ff99', '#66cc00', '#ffcccc', '#9999ff'),
        labels = c(
            'Steady state', 'SERPINB3 high', 'KRT24 high', 'CXCL1 high', 'Inflamed', 'KRT23 high'
        )
    ),
    Secretory = data.frame(
        subcluster2 = c('s1', 's2', 's3'),
        order = c(2, 1, 3),
        color = c('#6699cc', '#00ccff', '#ffcccc'),
        labels = c(
            'SERBINB3/PI3 high', 'Steady state', 'Inflamed'
        )
    )
)


split <- (
    split
    %>% imap(~add_subcluster2(.x, prefixes[[.y]]))
    %>% imap(~reorder_subcluster2(.x, orders[[.y]]))
    %>% imap(~add_subcluster2_text(.x, extras[[.y]]))
)
    
mkcolor <- function(df) {
    c <- df$color
    names(c) <- df$labels
    c
}
    
colors <- (
    extras
    %>% map(~arrange(.x, order))
    %>% map(mkcolor)
)    
```

# 4. Saving subclusters

```r
renamer <- function(cell_type) {
    newcolname <- paste(cell_type, 'subcluster', sep='_')
    function(colnames) colnames %>% str_replace('subcluster2_text', newcolname)
}

save_subclusters <- function(tbl, cell_type) (
    tbl
    %>% select(barcode, subcluster2_text)
    %>% rename_with(renamer(cell_type), all_of('subcluster2_text'))
    %>% write_csv(path=path_at('out', 'precomputed')(cell_type, 'subclusters', 'csv'))   
)

. <- (
    split
    %>% map(get_metadata)
    %>% imap(~save_subclusters(.x, .y))
)
```
