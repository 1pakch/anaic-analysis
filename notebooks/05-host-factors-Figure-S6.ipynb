```r
library(purrr)
library(tidyverse)
library(Seurat)
library(ggplot2)
library(scales)
```

```r
source('paths.R')
source('plotting.R')
source('misc.R')
```

# 1. Load Data

```r
# SCTransform must be recomputed as it is not properly persisted

d <- load_object(path_at('out', 'cache')('annotated', 'Robj')) %>% SCTransform(return.only.var.genes = FALSE)
```

```r
m <- d@meta.data
```

```r
hostFactors <- read_csv('in/host-factors/host-factors.csv')$gene
hostFactors
```

# 2. DE test between Secretory and Ciliated for host factors

```r
mask1 <- (m$cell_type %in% c('BdiS', 'Secretory', 'Ciliated'))
mask2 <- m$condition_text == 'Mock steady state'
```

```r
dd <- d[hostFactors, mask2]
```

```r
hf_de <- FindMarkers(dd,
            group.by='cell_type', ident.2='Secretory', ident.1='Ciliated',
            logfc.threshold=0.0, min.pct=0.5)
hf_de
```

# 3. Figure S6

```r
hf_de_genes <- rownames(hf_de)
dt <- t(dd[hf_de_genes,]@assays$SCT@scale.data) %>% as.data.frame
ct <- dd@meta.data %>% select(cell_type)
tt <- cbind(ct, dt)
ttl <- tt %>% pivot_longer(where(is.numeric), names_to='gene')
ttl$gene <- ttl$gene %>% factor(levels=hf_de_genes, order=TRUE)
```

```r
ps <- hf_de
ps$gene <- rownames(ps)
ps$asterisk <- if_else(ps$p_val_adj < 0.01, '*', '')
ps$p <- scales::pvalue(ps$p_val_adj, accuracy = 1e-3, add_p = TRUE)
ps$p <- sprintf('\n(%s)', ps$p)
ps$label <- paste0(ps$gene, ps$asterisk, ps$p)
pss <- ps$label
names(pss) <- ps$gene
pss
```

```r
display_host_genes <- c(
    'CLTA', 'EPS8', 'CTTN', 'EPN1',
    'HSP90AA1', 'HMGB1', 'CD151', 'CHD3',
    'TMPRSS4', 'RACK1', 'PTK2', 'ARHGAP21'
)
```

```r
labels <- as_labeller(pss)

options(repr.plot.width=8, repr.plot.height=3.5)

(
    ttl
    %>% filter(gene %in% display_host_genes)
    %>% mutate(gene = gene %>% factor(display_host_genes, ordered=TRUE))
    %>% arrange(gene)
    %>% ggplot(aes(y=cell_type, x=value, color=cell_type)) +
    geom_boxplot(notch=FALSE) +
    facet_wrap(vars(gene), ncol=4, scales='free_x', labeller=labels) +
    scale_y_discrete(limits=rev(c('Secretory', 'Ciliated'))) +
    scale_color_manual(values = cell_type_colors[c(3,4)]) +
    ylab('') + 
    xlab('Normalized expression') +
    theme(legend.position = 'none')
)

ggsave(path_at('out', 'figures')('Figure_S6', 'png'), width=8, height=3.5, dpi=250)
```

Figure S6. Expression of host factors involved in IV replication in steady state secretory and ciliated cells.
The figure shows the box-and-whisker plots of normalized gene expression. The x scale are Pearson residuals from SCTransform binomial regression model while p-values are from Mann-Whitney-Wilcoxon test with multiple testing adjustment (see Methods).
The boxes visualize the interquartile range of the respective distributions with the vertical lines corresponding to the medians. The lines on the left (right) of the boxes extend till the smallest (largest) observation or the first (third) quartile minus (plus) 1.5 times the IQR depending whichever is greater (smaller). The points outside this range are, arguably, outliers and visualized individually as points.
