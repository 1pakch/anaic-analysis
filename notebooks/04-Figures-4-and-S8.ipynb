# 2. Subcluster Abundances Plot (Figure 4)

```r
options(repr.plot.width=6, repr.plot.height=6)

label_percent_reverse = function(x) scales::label_percent(accuracy=1)(1-x)

make_data <- function(obj) (
    left_join(
        obj@meta.data %>% group_by(condition, subcluster) %>% count(name='n_subcluster'),
        obj@meta.data %>% group_by(condition) %>% count(name='n_condition'))
    %>% mutate(pct=n_subcluster/n_condition)
    %>% mutate(text=sprintf('%.0f%%', pct*100))
    %>% mutate(text=str_replace(text, '^0%$', ''))
)
    
plot <- function(data, k) (
    data
    %>% ggplot(aes(x=condition, y=pct, fill=subcluster, label=text))
    + geom_col()
    + geom_text(size=2, position = position_stack(vjust = 0.5))
    + scale_fill_manual(k, values=subcluster_metas[[k]]$color)
    + coord_flip()
    + scale_x_discrete(limits=condition_levels)
    + scale_y_reverse(label=label_percent_reverse)
    + theme(
        axis.title = element_blank(),
        axis.text = element_text(size=7),
        legend.key.size = unit(0.4, 'cm'),
        legend.key.height = unit(0.4, 'cm'),
        legend.key.width = unit(0.4, 'cm'),
        legend.title = element_text(size=7),
        legend.text = element_text(size=7))
)

plots <- (
    split
    %>% map(make_data)
    %>% imap(~plot(.x, .y))
)
    

options(repr.plot.width=8, repr.plot.height=7)

p <- patchwork::wrap_plots(plots, ncol=1, nrow=4)  & theme(legend.justification = "left")
ggsave(path_at('out', 'figures')('Figure_4', 'png'), width=6, height=6, dpi=250)
print(p)
```

Figure 4. Relative percentages of subclusters identified in the main HAE cell types across conditions: basal cells A), basal cells differentiating into secretory B), secretory cells C), and, ciliated cells D). Conditions as specified in Figure 2C.

# 3. Heatmaps (Figure S8)

## 3.1 Prepare x-axis metadata (cells)

```r
cells.group.vars <- c('condition', 'subcluster')

cells.meta <- (
    split
    %>% map(~.x@meta.data %>% as_tibble(rownames='barcode'))
    %>% map(select, barcode, condition, subcluster, nCount_RNA_all_genes, pct_viral, viral_load_class)
    %>% map(arrange, condition, subcluster, pct_viral)
    %>% map(~prepare_axis_meta(.x, cells.group.vars, relative_break_sizes()))
    %>% map(~mutate(.x, `log(#umi)`=log10(nCount_RNA_all_genes) - 4))
    %>% map(~mutate(.x, `% viral` = pct_viral))
    %>% map(~rename(.x, pos_x=pos))
    %>% map(~left_join(.x, vlc_meta %>% select(viral_load_class, color_fill) %>% rename(vlc_color_fill=color_fill)))
)
```

## 3.2 Prepare subplots (shown on top of the heatmap)

```r
options(repr.plot.width=12, repr.plot.height=4)

vlc_barplot <- function(axis.meta) (
    ggplot(data=axis.meta)
    + geom_raster(aes(x=pos_x, y='viral load'), fill=axis.meta$vlc_color_fill)
    + scale_x_discrete(limits=axis.meta %>% pull(pos_x) %>% factor())
    + theme(
        legend.position='none',
        axis.title=element_blank(),
        axis.text.x=element_blank(),
        axis.ticks=element_blank())
)

groups.barplot3 <- function(axis.meta, pos_var, group_vars, level, colors) (
    ggplot(data=axis.meta)
    + geom_raster(aes(x=get(pos_var), y=group_vars[[level]], fill=get(group_vars[[level]])))
    + scale_x_discrete(limits=axis.meta %>% pull(pos_var) %>% factor())
    + scale_fill_manual(name=NULL, values=colors, breaks=names(colors), guide='legend', drop=FALSE, na.translate=FALSE)
    + theme(
        legend.position='top',
        axis.title=element_blank(),
        axis.text.x=element_blank(),
        axis.ticks=element_blank())
)



condition.plots <- (
    cells.meta
    %>% map(~groups.barplot2(.x, 'pos_x', cells.group.vars, 1, name=NULL) + theme(legend.title = element_blank()))
)

subcluster.plots <- (
    cells.meta
    %>% imap(~groups.barplot3(.x, 'pos_x', cells.group.vars, 2, colors=subcluster_metas[[.y]]$color))
)

vlc.plots <- (
    cells.meta
    %>% map(vlc_barplot)
)

num_count.plots <- cells.meta %>% map(level.barplot, 'pos_x', 'log(#umi)', color='green')
#pct_viral.plots <- cells.meta %>% map(level.barplot, 'pos_x', '% viral', color='red')

condition.plots[[1]] + subcluster.plots[[1]] + num_count.plots[[1]] + vlc.plots[[1]] + patchwork::plot_layout(ncol=1)
```

## 3.3 Function wrapping a proper heatmap into a full plot

```r
mk_plot <- function(k, out_path, heatmap.plots, genes.meta, stretch_x=1, stretch_y=1) {
    hm_size <- heatmap_size(genes.meta[[k]], cells.meta[[k]], stretch_x=stretch_x, stretch_y=stretch_y)
    heights <- c(
        #title_height(),
        legend_height(),
        top_subplot_height(),
        top_subplot_height(),
        top_subplot_height(),
        top_subplot_height(),
        top_subplot_spacer_height(),
        hm_size[[2]])
    
    total_width <- 12
    total_height <- sum(heights)
    
    options(repr.plot.width=total_width, repr.plot.height=total_height)
    cat(heights)

    p <- (
        patchwork::guide_area()
        + (condition.plots[[k]] %>% smaller_legend)
        + (subcluster.plots[[k]] %>% smaller_legend)
        + (num_count.plots[[k]] %>% smaller_legend)
        + (vlc.plots[[k]] %>% smaller_legend)
        + patchwork::plot_spacer()
        + (heatmap.plots[[k]] %>% smaller_legend)
        + patchwork::plot_layout(guides='collect', nrow=length(heights), ncol=1, heights=heights %>% as_unit('inches'))
        & theme(plot.margin=margin(t=0, b=0, r=1.5, l=0, unit='pt'), legend.position='top')
    )
    ggsave(out_path(k), width=total_width, height=total_height+0.25)
    p
}
```

## 3.4 Create y-axis metadata (genes) from manually created gene lists

```r
read_curated_genes <- function(path) (
    read_csv(file=path, col_types=cols(gene=col_character()))
    %>% mutate(gene=toupper(gene))
    %>% mutate(gene=str_replace(gene, 'ORF', 'orf'))
    %>% mutate(gene_group=1)
    %>% prepare_axis_meta('gene_group', 1)
    %>% rename(pos_y=pos)
    %>% mutate(pos_y=factor(pos_y, levels=rev(pos_y)))
)

curated_genes.meta <- (
    split
    %>% imap(~paste('in/curated_genes/', .y, '-v5.csv', sep=''))
    %>% map(read_curated_genes)
)
```

## 3.5 Create heatmaps

```r
heatmaps.curated_genes <- (
    split
    %>% imap(~get.plot.cells.data(.x, curated_genes.meta[[.y]], cells.meta[[.y]], exmax=2.5, exmin=-2.5))
    %>% imap(~make.heatmap(.x, curated_genes.meta[[.y]], cells.meta[[.y]], limits=c(-2.5, 2.5)))
)
```
