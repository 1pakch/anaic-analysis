# 2. t-SNE scatterplots

## 2.A t-SNE colored by global clusters and conditions

```r
# Plotting functions

scatterplot_theme <- theme(
    axis.line = element_blank(),
    axis.ticks = element_blank(),
    axis.title = element_blank(),
    axis.text = element_blank(),
    legend.text = element_text(size=15)
)

global_clusters_scatterplot <- function(d, reduction_name) (
    DimPlot(d, reduction=reduction_name, group.by='global_clusters', shuffle=TRUE, pt.size=0.01)
    + global_clusters_scale_color
    + ggtitle(NULL)
    + scatterplot_theme
    + theme(legend.text=element_text(size=rel(0.75)))
)

conditions_scatterplot <- function(d, reduction_name) (
    DimPlot(d, reduction=reduction_name, group.by='condition', shuffle=TRUE, pt.size=0.01)
    #+ aes(color = condition)
    + scale_color_manual(values=condition_colors)
    + ggtitle(NULL)
    + scatterplot_theme
    #+ guides(legend=guide_legend(override.aes = list(width=6)))
    + theme(legend.text=element_text(size=rel(0.75)))
)
```

```r
options(repr.plot.width=8, repr.plot.height=4)
tsne_gc <- global_clusters_scatterplot(d, 'tsne_pcs')
tsne_condition <- conditions_scatterplot(d, 'tsne_pcs')
#tsne_gc
#tsne_condition\
tsne_condition
```

# 3. Abundances of global clusters / cell types

```r
plot_freqs <- function(freqs) (
    freqs
    %>% ggplot(aes(x=condition, y=freq, fill=global_clusters))
    + geom_col(position = position_stack(reverse=TRUE))
    + geom_text(
        aes(label = percent(freq, accuracy=1)),
        position = position_stack(vjust=0.5, reverse=TRUE),
        size = 3,
        color = 'white')
    + global_clusters_scale_fill
    + coord_flip()
    + scale_x_discrete(limits=rev(levels(d@meta.data$condition)))
    + scale_y_continuous(limits=c(0, 1), expand = expansion(mult = 0.025, add = 0))
    + theme(
        axis.title=element_blank(),
        legend.title=element_blank(),
        axis.text.x = element_blank(),
        axis.text.y = element_text(size=11),
        axis.ticks = element_blank(),
        legend.position = 'none',
        panel.background = element_blank()
    )
)
```

```r
options(repr.plot.width=6, repr.plot.height=2)

glc <- (
    d@meta.data
    %>% filter(global_clusters != 'Undefined')
    %>% select(condition, global_clusters)
)

conditions <- glc %>% group_by(condition) %>% count(name='n_condition')
clusters <- glc %>% group_by(condition, global_clusters) %>% count(name='n')
freqs <- left_join(clusters, conditions) %>% mutate(freq=n/n_condition)

ct_freqs_plot <- plot_freqs(freqs)
ct_freqs_plot
```

```r
options(repr.plot.width=6, repr.plot.height=2)

glc <- (
    d@meta.data
    %>% select(condition, global_clusters)
)

conditions <- glc %>% group_by(condition) %>% count(name='n_condition')
clusters <- glc %>% group_by(condition, global_clusters) %>% count(name='n')
freqs <- left_join(clusters, conditions) %>% mutate(freq=n/n_condition)

gc_freqs_plot <- plot_freqs(freqs)
gc_freqs_plot
```

```r
options(repr.plot.width=8, repr.plot.height=2)
freq_plots <- (gc_freqs_plot + (ct_freqs_plot + theme(axis.text=element_blank(), axis.ticks=element_blank())))
freq_plots
```

# 4. Heatmaps

```r
genes.meta <- (
    hallmark2
    %>% mutate(cell_type=1)
    %>% prepare_axis_meta(c('cell_type'), absolute_break_sizes(1))
    %>% rename(pos_y=pos)
    %>% mutate(pos_y=factor(pos_y, levels=rev(pos_y)))
)
```

```r
cells.meta <- (
    d@meta.data
    %>% as_tibble(rownames='barcode')
    %>% select(barcode, global_clusters, nCount_RNA_all_genes, pct_viral)
    %>% arrange(global_clusters, pct_viral)
    %>% prepare_axis_meta(c('global_clusters'), relative_break_sizes())
    %>% mutate(`log(reads)`=log10(nCount_RNA_all_genes) - 4)
    %>% mutate(`% viral` = pct_viral)
    %>% rename(pos_x=pos)
)
```

```r
global_clusters.plot <- cells.meta %>% groups.barplot('pos_x', 'global_clusters', 1, palette=2)
num_count.plot <- cells.meta %>% level.barplot('pos_x', 'log(reads)', color='green')
pct_viral.plot <- cells.meta %>% level.barplot('pos_x', '% viral', color='red')
```

```r
heatmap <- (
    make.heatmap(get.plot.cells.data(d, genes.meta, cells.meta), genes.meta, cells.meta)
)
```

```r
mk_plot <- function(cells.meta, filename, stretch_x=1, stretch_y=1) {
    hm_size <- heatmap_size(genes.meta, cells.meta, stretch_x=stretch_x, stretch_y=stretch_y)
    heights <- c(
        #title_height(),
        legend_height(),
        top_subplot_height(),
        top_subplot_height(),
        top_subplot_height(),
        top_subplot_spacer_height(),
        hm_size[[2]])
    
    total_width <- 12
    total_height <- sum(heights)
    
    options(repr.plot.width=total_width, repr.plot.height=total_height)
    cat(heights)

    p <- (
        patchwork::guide_area()
        + (global_clusters.plot %>% smaller_legend)
        + (num_count.plot %>% smaller_legend)
        + (pct_viral.plot %>% smaller_legend)
        + patchwork::plot_spacer()
        + (heatmap %>% smaller_legend)
        + patchwork::plot_layout(guides='collect', nrow=length(heights), ncol=1, heights=heights %>% as_unit('inches'))
        & theme(plot.margin=margin(t=0, b=0, r=1.5, l=0, unit='pt'), legend.position='top')
    )
    p
}
```

```r
heatmap_plot <- mk_plot(cells.meta, 'all')
```

```r
options(repr.plot.width=8, repr.plot.height=4)

heatmap_plot
```
